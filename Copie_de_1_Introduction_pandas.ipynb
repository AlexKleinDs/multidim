# 4. Indiçage (_indexing_)

## a. Introduction

L'_indiçage_ en python se fait dans les crochets. Avec `pandas` on peut indicer des données à la fois avec des numéros, mais aussi avec des chaînes de caractères.
Vous trouverez de nombreuses pages de [documentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html) sur le sujet.

Le but de cette section est de comprendre la logique sous-tendant l'indiçage en `pandas`. C'est un outil très puissant pour ce genre de chose, mais ce n'est pas trivial de tout comprendre au début !

## b. Un `DataFrame` comme une liste de `Series`

Formellement, le plus simple pour commencer est de comprendre un objet `DataFrame` comme une collections de colonnes. Chacune d'elle est simplement une suite de données, et il s'agit d'un objet de type [`Series`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html).

Losqu'on extrait des colonnes d'un `DataFrame`, il y a deux solutions:
* On sélectionne plusieurs colonnes d'un coup, cela produit une nouvelle `DataFrame`, qu'on peut donc manipuler comme la structure initiale.
* On sélectionne une seule colonne, et on obtient alors un objet de type [`Series`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html), il s'agit de la suite des données de la colonne en question

````
months = ['Janvier', 'Février', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet',
          'Août', 'Septembre', 'Octobre', 'Novembre', 'Décembre']
# Le type de `data` est `DataFrame`
print('type de `data`\n  ',type(data))

# tout comme le type de `data[months]`
print('type de `data[months]`\n  ', type(data[months]))

# et le type d'une sélection de colonnes
print("type de `data[['Janvier', 'Mars']]`\n  ", type(data[['Janvier', 'Mars']]))

# Par contre, le type de data['Janvier'] est différent:
print("\ntype de `data['Janvier']`\n  ", type(data['Janvier']))
````

Notons ici qu'accéder à la `Series` d'une colonne donnée (e.g. 'Mars') peut se faire soit par `data['Mars']` soit par `data.Mars`:

````
data.Mars
````

````
# enregistrons pour plus tard les noms des mois, ce sont les noms des 12 premières colonnes de la table
# convertissons ces noms en une simple liste python 
months = data.columns[:12].tolist()

# par exemple, affichons juste les colonnes donnant les températures mensuelles. C'est un `DataFrame`
data[months].head()
````

> Cette distinction entre la sélection d'une ou plusieurs colonnes explique pourquoi:
   * `data['Janvier']['Amsterdam']` _fonctionne_: on accède à l'entrée 'Amsterdam' de la `Series`: `data['Janvier']`
   * `data[months]['Amsterdam']` _ne fonctionne pas_: on cherche en faisant ça à accéder à la __colonne__ 'Amsterdam' de la `DataFrame` créee, qui n'existe pas: 'Amsterdam' est une ligne !


````
# La ligne suivante fonctionne
print('Température en Janvier à Amsterdam: ', data['Janvier']['Amsterdam'])
````

````
# Mais la ligne suivante ne fonctionne pas
print('Température en Janvier à Amsterdam: ', data[['Janvier', 'Mars']]['Amsterdam'])
````

---
Pour l'instant, nous avons vu le moyen de:
* Sélectionner des colonnes entières

  $\Rightarrow$ par exemple:
    * `data[months]` (c'est un `DataFrame`)
    * `data.Février` (c'est une `Series`).
* Sélectionner des éléments particuliers d'une `Series`

  $\Rightarrow$ par exemple `data.Mars['Amsterdam']`

---

Notons qu'on peut facilement sélectionner des lignes entières avec cette logique, en partant de la base transposée

````
data.T['Dublin']
````

## c. Slicing en `pandas`

Il est possible en `pandas` de faire des sélections par _tranches_ (`slicing`). Cela se fait, toujours entre crochets en utilisant la syntaxe `data[début:fin]`

> __Attention__   
  Par convention, le _slicing_ se fait sur les __lignes__ d'une DataFrame
 

````
# on peut faire des slices avec un index numérique... ici de la sixième à là 10ème ligne
# par convention, la dernière position n'est pas incluse dans ce cas
print(data[5:10])

# ou bien des slices avec les labels alphanumériques ! (ici les noms des villes).
# par convention, toutes les entrées sont retournées dans ce cas.
print(data['Budapest':'Kiev'])

````

````
# Encore une fois, on peut sélectionner des slices de colonne, simplement en travaillant sur la table transposée
data.T['Avril':'Août'].T.head()
````

## d. Simplifier l'indexing: `loc`, `iloc` et `at`

Tous ces comportements, bien que logique, peuvent paraître un peu lourds ! C'est pour cette raison qu'il existe deux vues plus pratiques pour accéder aux données, appelées `loc`, `iloc`.

### Création de sous-tables par labels: `loc`

De manière à rendre l'accès aux données un peu plus naturel, tout `DataFrame` a un attribut `loc`, qui permet d'accéder aux données plus simplement, en prenant à la fois des lignes et des colonnes:

````
# accès à un seul élément (produit un simple scalaire)
print("La température d'Amsterdam en février est", data.loc['Amsterdam', 'Février'])

# accès à une sélection au sein d'une seule colonne (produit un objet `Series`)
print("\nLes températures en Août des villes alphabétiquement entre Berlin et Paris sont\n",
      data.loc['Berlin':'Paris', 'Août'])

# extraction directe d'une sous table (produit un DataFrame)
print("\nExtraction directe d'une sous-table")
data.loc['Paris':'Sofia', 'Juin':'Octobre']
````

### Création de sous tables par indexs numériques: `iloc`

Il arrive qu'on souhaite plutôt accéder aux élments de la table par leur numéro de ligne et de colonne, c'est exactement ce que fait `iloc`, qui permet de construire facilement des sous-tables  

````
data.iloc[5:10, :12]
````

````
data.iloc[:, 3]
````

### Accéder à un seul élément: `at`

Pour finir ce bestiaire des manières d'accéder aux données de la table, la technique la plus simple pour obtenir une entrée est la vue `at`, à laquelle on peut accéder avec une seule ligne et colonne

````
# ou même utiliser le fait qu'on a nommé nos lignes (c'est "l'index").

# `at` est une vue des données pratique pour accéder à une entrée de la table
print('La température d\'Amsterdam en juin est: ', data.at['Amsterdam', 'Juin'])
````

## b. Pivotage

Même si le format tabulaire (individus, grandeur) est pratique à visualiser pour un individu, c'est très rarement le format idéal pour analyser les données, surtout lorsqu'elles sont _catégorielles_, c'est à dire qu'elles prennent un nombre fini de valeurs.

Par exemple, dans notre cas, une manière alternative de représenter les données serait de se concentrer sur _chaque température du tableau_ et se demander: 
* À quelle ville elle correspond ?
* Quelle est le mois correspondant ?

Ainsi, en partant de l'exemple suivant:

| Ville | Janvier | Août |
|-------|---------|------|
|Paris |  3.7 | 18.7|
|Rome | 7.1 | 24.1|

On voudrait arriver à:

|Ville | Mois | Température|
|------|------|------------|
|Paris | Janvier | 3.7|
|Paris | Août | 18.7|
|Rome | Janvier | 7.1|
|Rome | Août | 24.1|

En effet, ce genre de représentation rend beaucoup plus facile certains traitements, comme on le verra plus tard. On appelle ce genre de représentation alternative un [__pivot__](https://blog-postgresql.verite.pro/2018/06/02/crosstab-pivot.html) des données, et il est important que vous sachiez comment faire.

`pandas` offre la possibilité de facilement réaliser de tels traitements. Vous trouverez une explication assez graphique des différentes opérations [ici](https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html).


````
data[:5]
````

````
# dans notre exemple, on a plutôt besoin d'un `melt`
data.reset_index().melt(id_vars='Ville', value_vars=months, var_name='Mois', value_name='Température')
````

# 6. Descriptions quantitatives des données

## a. `describe`

On a vu comment accéder aux données se fait en `pandas`, et quelques fonctionnalités pour les réorganiser.

Voyons maintenant comment commencer à analyser les données.

Une manière très simple d'avoir un résumé est d'utiliser la fonction `describe`

````
# describe the data and store the result
description = data[months].describe()

# display the description
description
````

Comme on le voit, la description est elle-même un objet pandas (une `DataFrame`). On peut donc la manipuler de la même manière:

````
print(' La température moyenne en mai est : ', description.at['mean', 'Mai'])
print('Les températures maximales par mois sont:\n', description.loc['max'])
````

## b. Agrégation

On pourrait vouloir des analyses plus fines des données.
Par exemple, on pourrait vouloir la valeur moyenne des températures pour chaque mois et chaque région.
Ce genre de traitement s'appelle de __l'agrégation__ et se fait très simplement en `pandas` avec la fonction `pivot_table`

````
aggregated = data.pivot_table(values=months, index='Région')
aggregated
````

````
# on remarque que les différentes colonnes ne sont pas bien ordonnées.
# on peut réarranger cette table en ordonnant les colonnes selon l'ordre du mois dans l'année
aggregated[months]
````

## a. Avec pandas

Pour un troisième exemple, on pourrait afficher une boîte à moustache (boxplot) des températures selon la région

## b. Avec `seaborn`

Même si `pandas` est un package très puissant et possède certaines fonctions d'affichage, ce n'est pas sa fonction fondamentale d'offrir des fonctionalités de visualisation.

C'est par contre à cette fin qu'est développé un autre package très utile: [`seaborn`](https://seaborn.pydata.org/). En plus, `seaborn` est __conçu pour s'interfacer avec `pandas`__. En pratique, `seaborn` est une surcouche de la très puissante (mais moins facile à utiliser) bibliothèque [`matplotlib`](https://matplotlib.org/).

Voyons comment réaliser quelques affichages des données. Vous êtes invités à vous même regarder la [galerie d'exemples](https://seaborn.pydata.org/examples/index.html), pour voir ce que vous voudriez pouvoir afficher

````
# on importe seaborn, et on règle tous ses paramètres esthétiques à leurs valeurs par défaut
import seaborn as sns
sns.set()

# premier exemple: un boxplot (boite à moustache) des températures par mois.

# on réarrange les données pour avoir toutes les températures "en vrac", avec leur région et mois correspondant
flattened = data.melt(id_vars='Région', value_vars=months, var_name='Mois', value_name='Température')
g = sns.boxplot(x='Mois', y='Température', data=flattened, hue='Région');

# on tourne les labels pour que ça soit plus joli
for item in g.get_xticklabels():
    item.set_rotation(45)

# on agrandit la figure
g.get_figure().set_size_inches(10,5)
````

````
# Deuxième exemple: un affichage de la distribution jointe des températures sur deux mois donnés
# ainsi que les distributions marginales de chaque mois

sns.jointplot(x='Janvier', y='Août', kind='kde', shade_lowest=False, data=data);
````

````
# Troisième exemple: on peut visualiser des données multidimensionnelles avec un pairgrid plot.
# Ça correspond à la figure précédente, mais pour chaque couple de variables. On peut aussi par
# exemple faire cet affichage d'une couleur différente en fonction d'une variable explicative
# (ici "Région") 

# création du canvas de pairgrid
g = sns.PairGrid(data, diag_sharey=False, vars=months[::4], hue='Région')

# les termes non diagonaux sont traités avec kdeplot
g.map_offdiag(sns.kdeplot)

# pour les termes diagonaux, on affich la distribution marginale (histogramme)
g.map_diag(sns.kdeplot,lw=3, legend=False)

# on rajoute la légende et on agrandit la figure
g = g.add_legend()

# on agrandit la figure
g.fig.set_size_inches(15,10)

````

````
# Quatrième exemple: affichage des histogrammes des températures de chaque ville 
# on commence par réarranger les données
flattened = data.reset_index().melt(id_vars = ['Ville'], value_vars=months, var_name='Mois', value_name='Température')

# affichage du violinplot
g = sns.violinplot(data=flattened, x='Ville', y='Température')

# on tourne les labels pour que ça rentre
for item in g.get_xticklabels():
    item.set_rotation(45)

# on agrandit la figure
g.get_figure().set_size_inches(20,5)
````
